## Preparation

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('./db/demo_lecture3.db')
```

## Basic SELECT

```python
sqlstr = '''
SELECT studentName FROM student;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT DISTINCT studentName FROM student;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentName FROM student LIMIT 2;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, score FROM score ORDER BY score LIMIT 5;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, score FROM score ORDER BY score DESC LIMIT 5;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, term, score FROM score ORDER BY term, score DESC LIMIT 10;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, score FROM score WHERE score BETWEEN 85 AND 90;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, studentName, native FROM student WHERE studentName = '王红';
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, studentName, native FROM student WHERE studentName != '王红';
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, studentName, nation FROM student WHERE nation = '汉族' AND sex = '男';
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, studentName, nation FROM student WHERE nation = '傣族' OR nation = '蒙古族' ORDER BY nation;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, studentName, nation FROM student WHERE nation IN ('傣族', '蒙古族') ORDER BY nation;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, studentName, nation FROM student WHERE nation NOT IN ('傣族', '蒙古族') ORDER BY nation;
'''
pd.read_sql(sqlstr, conn)
```

## LIKE & REGEXP

```python
sqlstr = '''
SELECT studentNo, studentName, native FROM student WHERE studentNo LIKE '80%';
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentNo, studentName, native FROM student WHERE studentNo LIKE '70100_';
'''
pd.read_sql(sqlstr, conn)
```

## JOIN

```python
sqlstr = '''
SELECT studentName, grade, nation FROM student, class WHERE student.classNo = class.classNo;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentName, grade, nation FROM student INNER JOIN class ON student.classNo = class.classNo;
'''
pd.read_sql(sqlstr, conn)
```

```python
sqlstr = '''
SELECT studentName, grade, nation FROM student INNER JOIN class ON student.classNo = class.classNo WHERE class.grade = '2008' AND student.nation = '蒙古族';
'''
pd.read_sql(sqlstr, conn)
```

## VIEW

```python
conn.execute('drop view if exists student_class;')
sqlstr = '''
CREATE VIEW student_class AS SELECT studentNo, studentName, sex, birthday, native,nation, class.* FROM student INNER JOIN class ON student.classNo = class.classNo;
'''
conn.execute(sqlstr)
sqlstr = '''
SELECT * FROM student_class;
'''
pd.read_sql(sqlstr, conn)
```
